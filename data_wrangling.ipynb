```python
nvalues = data.isna().sum()
```

```python
nvalues[nvalues.gt(0)]/len(data)
```

```python
#all columns have more that 50% null values 
#last two columns almost 100% -> drop those columns

nvalues_ratios = nvalues[nvalues.gt(0)]/len(data)
nvalues_ratios.loc[nvalues_ratios.gt(0.9)].index

data2 = data.drop(columns=nvalues_ratios.loc[nvalues_ratios.gt(0.9)].index).copy()

data2
```

```python
#dropping rows with all null values 
data3 = data2.dropna(how = 'all')
```
