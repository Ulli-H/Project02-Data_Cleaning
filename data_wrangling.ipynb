```python
#checking for missing values
#with boolean map
data.isna()
```

```python
#calculating sum of boolean map and assigning to variable nvalues
nvalues = data.isna().sum()
```

```python
#calculating percentages of nvalues among all values in columns
nvalues[nvalues.gt(0)]/len(data)
```

```python
#all columns have more that 50% null values 
#last two columns almost 100% -> drop those columns

#assign percentages to the variable nvalues_ratios
nvalues_ratios = nvalues[nvalues.gt(0)]/len(data)

#filter for percentages over 90%
nvalues_ratios.loc[nvalues_ratios.gt(0.9)].index

#dropping the two columns
data2 = data.drop(columns=nvalues_ratios.loc[nvalues_ratios.gt(0.9)].index).copy()

data2
```

```python
#dropping rows with all null values 
data3 = data2.dropna(how = 'all')
```

```python
data3
```

```python
#down to 8703 rows
#but what about rows with null values and a 0 as case number 

#data3.loc[data3['Case Number']== 0] -> doesn't work since column values 
#are type object, not numerical
data3.dtypes
```

```python
print(set(data3['Case Number']))
```

```python
#replacing all . and - in column Case Number with empty string 

data3['Case Number'] = data3.loc[:,'Case Number'].str.replace('.', '').replace('-','')
print(set(data3['Case Number']))
```

```python
#trying to replace the non-digits with empty string and than converting dtype of the column to integer - didn't work
data3['Case Number'] = data3['Case Number'].str.replace(r'\D+', '').astype('int')

#use to numeric instead
```

```python
#filling the na in column Case Number with 0, to filter later for 0 and drop these rows 

data3['Case Number'] = data3['Case Number'].fillna(0)
```

```python
#replacing the non-digits with empty string 
data3['Case Number'] = data3['Case Number'].str.replace(r'\D+', '')
print(set(data3['Case Number']))
```

```python
#replacing 'Nan' in column Case Number with empty string 
data3['Case Number'] = data3['Case Number'].str.replace( 'NaN' , '')
data3
```
